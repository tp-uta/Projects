```python
import string
from pandas import *
import numpy as np

#Representation of players and an empty space
player_1 = 1
player_2 = 2
empty = 0

#Drawing Representation of above
player_1_piece = "X"
player_2_piece = "O"
empty_space = " "

space_character = {player_1: player_1_piece,
                  player_2: player_2_piece,
                  empty: empty_space}

def empty_board(n):
    temp = [[0]*n for i in range(n)]
    return temp

def make_move(arr, player, location):
    values = list(string.ascii_uppercase)
    move_dictionary = {}
    for i in range(26):
        move_dictionary[values[i]] = i
    location = list(location)
    if len(location) == 2 and location[0] in string.ascii_uppercase[0:len(arr)] and location[1] in string.digits:
        letter_location = move_dictionary[location[0]]
        number_location = int(location[1]) - 1
    else:
        print("Error: Invalid Input")
        return arr, False
    if letter_location not in range(len(arr)) or number_location not in range(len(arr)):
        print("Error: Spot is out of bounds")
        return arr, False
    elif arr[number_location][letter_location] == empty:
        arr[number_location][letter_location] = player
        return arr, True
    else:
        print("Error: Spot is already taken")
        return arr, False

def check_move(arr, player_num, valid):
    temp = arr
    while(valid == False):
        player_move = input("Invalid move. Enter another move: ")
        arr, valid = make_move(temp, player_num, player_move)
    return arr

def board(arr, count_move):
    x = list(string.ascii_uppercase)
    print(end = "    ")
    for i in range(len(arr)):
        print(f"{x[i]}", end = "   ")
    print("\n", end = "")
    print("   ---", end = "")
    for i in range(len(arr)):
        print(" ---"*(len(arr)-1))
        print(f"{i+1} |", end = "")
        for j in range(len(arr)):
            print(f" {space_character[arr[i][j]]} |", end = "")
        print()
        print("   ---", end = "")
    print(" ---"*(len(arr)-1))
    return f"Move number: {count_move}"

def diagonal(arr):
    temp = []
    count = 0
    for i in range(len(arr)):
        temp.append(arr[i][i])
    for i in range(len(temp)-1):
        if temp[i] == temp[i+1] and temp[i] != 0:
            count += 1
    if count == len(temp)-1:
        return 1, temp[0]
    else:
        return -1, False

def other_diagonal(arr):
    temp = []
    count = 0
    for i in range(len(arr)):
        temp.append(arr[i][len(arr)-1-i])
    for i in range(len(temp)-1):
        if temp[i] == temp[i+1] and temp[i] != 0:
            count += 1
    if count == len(temp)-1:
        return 1, temp[0]
    else:
        return -1, False

def horizontal(arr):
    for i in range(len(arr)):
        temp = []
        count = 0
        for j in range(len(arr)):
            temp.append(arr[i][j])
        for k in range(len(temp)-1):
            if temp[k] == temp[k+1] and temp[k] != 0:
                count += 1
        if count == len(temp)-1:
            return 1, temp[0]
    else:
        return -1, False
        
def vertical(arr):
    for i in range(len(arr)):
        temp = []
        count = 0
        for j in range(len(arr)):
            temp.append(arr[j][i])
        for k in range(len(temp)-1):
            if temp[k] == temp[k+1] and temp[k] != 0:
                count += 1
        if count == len(temp)-1:
            return 1, temp[0]
    else:
        return -1, False

def check_board_win(arr):
    x, temp = horizontal(arr)
    if x == -1:
        x, temp = vertical(arr)
    if x == -1:
        x, temp = other_diagonal(arr)
    if x == -1:
        x, temp = diagonal(arr)
    if x == 1:
        return 1, temp
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i][j] == 0:
                return -1, False
    return 0, False

def tictactoe_game():
    n = int(input("What size board? "))
    board_array = empty_board(n)
    num_moves = 0
    print(board(board_array, num_moves+1))
    player_num = player_1
    game_state = -1
    while(game_state == -1):
        if num_moves & 1:
            current_player = player_2
        else:
            current_player = player_1
        num_moves += 1
        print(f"Player {current_player} is making a move...")
        curr_move = input("Enter a move here: ")
        board_array, valid = make_move(board_array, current_player, curr_move)
        board_array = check_move(board_array, current_player, valid)
        game_state, state_num = check_board_win(board_array)
        print(board(board_array, num_moves+1))
        if game_state == 1:
            if state_num == player_1:
                print("Player 1 has won!")
            elif state_num == player_2:
                print("Player 2 has won!")
            else:
                print("???")
        elif game_state == 0:
            print("Draw. No one has won")
```
